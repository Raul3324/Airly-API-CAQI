```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
```

```python
# Citim datele
date_adunate = pd.read_csv("date_adunate.csv")
```

```python
plt.figure(figsize=(12,6)) # marimea graficului
graf = sns.scatterplot(data=date_adunate, x='ora', y='index', hue='dataset', legend=True) #  datele
sns.move_legend(graf, "lower right") # mutam legenda sa nu ascunda ceva
plt.xticks(range(24)) # valori pe axa x din 1 in 1, de la 0 la 23
for i in range(24):
    plt.axvline(i) # linii verticale in dreptul fiecarei valori de pe axa x
```

```python
g = sns.FacetGrid(date_adunate, col='weekday', hue='dataset', col_order=['Luni' ,'Marti', 'Miercuri', 'Joi', 'Vineri', 'Sambata', 'Duminica'])
g.map(sns.scatterplot, 'ora', 'index')
g.add_legend();
```
